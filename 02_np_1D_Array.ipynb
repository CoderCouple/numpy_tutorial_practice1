# Numpy 1-D basic array creation

```python
import numpy as np
```

```python
# 1-D array
array1 = np.arange(3)
print(array1)
```

```python
# 1-D array
array2 = np.arange(-2, 5)
print(array2)
```

```python
array3 = np.arange(0,11,2)
print(array3)
```

```python
array4 = np.arange(-1,-4,-1)
print(array4)
```

```python
array5 = np.arange(0.1,0.3,0.1)
print(array5)
```

```python
array6 = np.linspace(0.1,0.3,3)
print(array6)
```
