# Numpy 2D Arrays

```python
import numpy as np
```

```python
identity_3X3 = np.eye(3)
print(identity_3X3)
```

```python
indentity_3X4 = np.eye(3,4)
print(indentity_3X4)
```

```python
dialgonal_2d= np.diag([1,2,3,4,5])
print(dialgonal_2d)
```

```python
diag_array = np.array([
 [1, 0, 0, 0, 0],
 [0, 2, 0, 0, 0],
 [0, 0, 3, 0, 0],
 [0, 0, 0, 4, 0],
 [0, 0, 0, 0, 5]
])
diag_elements = np.diag(diag_array)
print(diag_elements)
```
